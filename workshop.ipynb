```
#loading matplotlib libarary
import matplotlib.pyplot as plt

```

```
x=[2,9,3,7]
y=[7,15,5,8]
plt.xlabel("distance") #x-axis
plt.ylabel("time") #y-axis
plt.plot(x,y,label="railway track") #this will draw a line
plt.scatter(x,y,s=200,label='railway station',marker='*',color='red')
plt.bar(x,y,label="height of railway station")#to draw bar chart
plt.grid(color="green") #divided into the boxes (grid)
plt.legend() #it will show label in the plot
plt.show() #it will show graph
```

```
#this will be another example of cricket data visulization
players=['dhoni','virat','rohit','jadeja']
scores=[120,114,150,80]

#now ploting it
plt.xlabel("players")
plt.ylabel("scores")
#plt.bar(players,scores,label="2020 socres of indian cricket team")
plt.pie(scores,labels=players,explode=(0.2,0,0,0),shadow=True,autopct='%1.1ff%%')
#plt.grid(color='green')
plt.legend()
plt.show()
```
